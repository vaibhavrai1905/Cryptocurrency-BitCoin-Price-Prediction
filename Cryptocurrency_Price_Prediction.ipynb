```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import datetime
from datetime import date, timedelta
today= date.today() #this helps to get todays data.
import yfinance as yf #This will help you collect the latest data each time you run this code:
```

```python
d1 = today.strftime("%Y-%m-%d") #"%Y-%m-%d" this is a format of year-month-day
                                # to format this datetime object into a string
```

```python
end_date = d1
```

```python
d2 = date.today() - timedelta(days=730) #A timedelta object displays the amount of time—in days, seconds, and microseconds
```

```python
d2 = d2.strftime("%Y-%m-%d")
```

```python
start_date = d2
```

```python
data=yf.download('BTC-USD', start=start_date, end=end_date, progress=False) #
```

```python
data['Date']= data.index
```

```python
data=data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.tail(10)
```

```python
'''In the above code, I have collected the latest data of Bitcoin prices for the past 730 days'''
```

```python
# Now let’s visualize the change in bitcoin prices till today by using a candlestick chart:
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                       open=data["Open"],
                                       high=data["High"],
                                       low=data["Low"],
                                       close=data["Close"])])
figure.update_layout(title = "Bitcoin Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
# Predicting the future prices of cryptocurrency is based on the problem of Time series analysis
```

```python
# The AutoTS library in Python is one of the best libraries for time series analysis.
```

```python
pip install autots
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
```

```python
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
